```python
import pandas as pd
```

```python
df=pd.read_csv('population.csv')
df
```

```python
dfspecific=df[df['Country Name']=='India']
dfspecific
```

# Visualize Population Of India 

```python
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(20,8))
```

```python
dfspecific.drop(['Country Name',"Country Code","Indicator Name","Indicator Code"],axis=1,inplace=True)
```

```python
df_transposed = dfspecific.T
df_transposed
```

```python
plt.bar(df_transposed.index,df_transposed[109],color='skyblue')
plt.title('Population Distribution Of India by Year')
plt.xlabel('Year')
plt.ylabel('Population')
plt.xticks(rotation=90, fontsize=10) 
plt.subplots_adjust(bottom=0.5)
# Show the plot

plt.grid(axis='y')
plt.tight_layout()
plt.show()



```

# Comparing Population of diff regions


```python
exclude=['Country Name',"Country Code","Indicator Name","Indicator Code"]
```

```python
df2=df.loc[:, df.columns.difference(exclude)]
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(12, 8)) 

# Plotting the population for the country at index 2
plt.plot(df2.columns, df2.loc[1], color='yellow', marker='o', label='Africa Eastern and Southern')
plt.plot(df2.columns, df2.loc[2], color='skyblue', marker='o', label='Afghanistan')  # Add label if needed
plt.plot(df2.columns, df2.loc[3], color='red', marker='o', label='Africa Western and Central')
plt.plot(df2.columns, df2.loc[4], color='green', marker='o', label='Angola')

plt.xlabel('Year')
plt.ylabel('Population')
plt.title('Population Growth by Country')
plt.legend()  # Show the legend if you added labels
plt.xticks(rotation=45)  # Rotate x-axis labels for better readability
plt.tight_layout()  # Adjust layout to fit everything
plt.legend(title='Countries')
plt.show()
```
